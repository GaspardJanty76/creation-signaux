```python
                                        
    
    
    
    
    
                                            # importation des bibliothèques 
    
    
    
    
    
    
import wave
import matplotlib.pyplot as plt
import pyaudio
import time
from math import *
from pylab import*
import numpy as np
PyAudio = pyaudio.PyAudio






                                            # Initialisation des variables

    
    



NF1= "Son1.wav"
NF2= "Son2.wav"
NF3= "Son3.wav"
NF4= "Son4.wav"
NF5= "Son5.wav"
NF6= "Son6.wav"
NF7= "Son7.wav"
NF8= "Son8.wav"
NF9= "Son9.wav"
NFNA= "SonNA.wav"


NC = 2 # nombre de canaux = 2 = stéréo
NO = 1 # taille d'un échantillon : 1 octet = 8 bits
fe = 44100 # fréquence d'échantillonnage
ng = 1 # niveau du son du canal de gauche
nd = 1 # niveau du son du canal de droite
t = 1 # durée
ag = 1 * ng # amplitude gauche 
ad = 1 * nd # amplitude droite
am = 1 # amplitude générale
nbE = t * fe # nombre d'échantillons (durée * fréquence d'échantillonnage)
pulsation = [] # création de la liste pulsation



                                                  # Définition
  




def son(): # initialisation de la définition
    
    
    print("                              création du son et du graph")
    
    
    s = wave.open(NomFichier, 'w') # création du fichier .wav
    parametres = (NC,NO,fe,nbE,'NONE','not compressed') # affiliation des paramètres
    s.setparams(parametres) # paramètrage de notre fichier son


    for i in range(0,nbE):
        vG = wave.struct.pack('B',int(128.0 + ag*math.sin(2.0*3.14*f*i/fe))) # écriture du canal audio gauche
        vD = wave.struct.pack('B',int(128.0 + ad*math.sin(2.0*3.14*f*i/fe))) # écriture du canal audio droit
        s.writeframes(vG + vD) # addition des deux canaux pour n'en fromer qu'un et écriture du son dans le fichier

    
    s.close() # fermeture du fichier son


    puls = 2 * 3.14 * f # calcul de la pulsation
    pulsation.append(puls) # ajout des resultats des pulsations dans la liste 'pulsation'
    def psi(x):
        return am*np.sin(puls*x) # fonction pour afficher la courbe du signal 

    # assignation des variables x et y
    x=np.linspace(0,t,nbE)
    y=psi(x)


    fig = plt.figure(figsize=(10,5)) # modification de la taille du graphique
    plt.grid(True) # affichage d'une grille dans le tableau
    plt.title('Graphique échantillon') # ajouter un titre au graphique
    plt.ylabel('Amplitude') # nommer y
    plt.xlabel('Temps') # nommer x
    plt.axis([0,0.002,-2,2]) # modification de la taille des valeurs
    plt.plot(x,y, '-') # assignation des valeurs dans le tableau 
    plt.show() # afficher le tableau
    
    
    print("                                     lecture du son")
    time.sleep(2) # attente de 2 secondes
    
    s = open(NomFichier,'rb') # ouverture en "lecture" du fichier son
    data = s.read() # lécture et implémentation des données du fichier son dans la variable data


    p = PyAudio() 
    # création d'un fichier pour lire le son
    stream = p.open(format = p.get_format_from_width(1), 
                    channels = 2, 
                    rate = 16000, 
                    output = True)


    # lancement de la lécture du fichier son
    stream.write(data)
    stream.stop_stream()
    stream.close()
    p.terminate()
   


def graph():
    print("                              création du son")
    fig = plt.figure(figsize=(10,5)) # modification de la taille du graphique
    plt.grid(True) # affichage d'une grille dans le tableau
    plt.title('Graphique échantillon') # ajouter un titre au graphique
    plt.ylabel('Amplitude') # nommer y
    plt.xlabel('Temps') # nommer x
    plt.axis([0,0.002,-2,2]) # modification de la taille des valeurs
    plt.plot(x,y, '-') # assignation des valeurs dans le tableau 
    plt.show() # afficher le tableau




        
        
                                                         # Fonction MAIN   
    

    

    
    
print("Choix de la fréquence du son 1")     
f = int(input()) # entrée de la frequence par l'utilisateur et affiliation à f
NomFichier = NF1  # affiliation de NF1 à NomFichier
    
son() # appel de la définition 'son'

print("Choix de la fréquence du son 2")    
f = int(input())
NomFichier = NF2

son()

print("Choix de la fréquence du son 3")    
f = int(input())
NomFichier = NF3
    
son()

print("Choix de la fréquence du son 4")    
f = int(input())
NomFichier = NF4 
    
son()

print("Choix de la fréquence du son 5")    
f = int(input())
NomFichier = NF5  
    
son()

print("Choix de la fréquence du son 6")    
f = int(input())
NomFichier = NF6

son()

print("Choix de la fréquence du son 7")    
f = int(input())
NomFichier = NF7
    
son()

print("Choix de la fréquence du son 8")    
f = int(input())
NomFichier = NF8 
    
son()

print("Choix de la fréquence du son 9")    
f = int(input())
NomFichier = NF9
    
son()

print("Choix de la fréquence du son non audible")    
f = int(input())
NomFichier = NFNA 
    
son()

print("choisiez le signal que vous voulez multiplier avec le signal non audible (1 à 9)")
n = int(input()) # entrée n saissie par l'utilisateur
if n > 0 & n < 10: # boucle si : si n est compris entre 0 et 10 (non inclus) faire :
    if n == 1: # si n = 1 faire :
        p1 = pulsation[0] # p1 prends la premiere valeur de la liste 'pulsation'
        p2 = pulsation[9] # p2 prends la derniere valeur de la liste 'pulsation'
        def psi(x):
            return am*np.sin(p1*x)+am*np.sin(2*p2*x) # fonction pour afficher la courbe du signal
        # affiliation de x et y
        x=np.linspace(0,t,nbE)
        y=psi(x)
        graph() # appel de la définition 'graph'
    if n == 2:
        p1 = pulsation[1]
        p2 = pulsation[9]
        def psi(x):
            return am*np.sin(p1*x)+am*np.sin(2*p2*x)
        x=np.linspace(0,t,nbE)
        y=psi(x)
        graph()
    if n == 3:
        p1 = pulsation[2]
        p2 = pulsation[9]
        def psi(x):
            return am*np.sin(p1*x)+am*np.sin(2*p2*x)
        x=np.linspace(0,t,nbE)
        y=psi(x)
        graph()
    if n == 4:
        p1 = pulsation[3]
        p2 = pulsation[9]
        def psi(x):
            return am*np.sin(p1*x)+am*np.sin(2*p2*x)
        x=np.linspace(0,t,nbE)
        y=psi(x)
        graph()
    if n == 5:
        p1 = pulsation[4]
        p2 = pulsation[9]
        def psi(x):
            return am*np.sin(p1*x)+am*np.sin(2*p2*x)
        x=np.linspace(0,t,nbE)
        y=psi(x)
        graph()
    if n == 6:
        p1 = pulsation[5]
        p2 = pulsation[9]
        def psi(x):
            return am*np.sin(p1*x)+am*np.sin(2*p2*x)
        x=np.linspace(0,t,nbE)
        y=psi(x)
        graph()
    if n == 7:
        p1 = pulsation[6]
        p2 = pulsation[9]
        def psi(x):
            return am*np.sin(p1*x)+am*np.sin(2*p2*x)
        x=np.linspace(0,t,nbE)
        y=psi(x)
        graph()
    if n == 8:
        p1 = pulsation[7]
        p2 = pulsation[9]
        def psi(x):
            return am*np.sin(p1*x)+am*np.sin(2*p2*x)
        x=np.linspace(0,t,nbE)
        y=psi(x)
        graph()
    if n == 9:
        p1 = pulsation[8]
        p2 = pulsation[9]
        def psi(x):
            return am*np.sin(p1*x)+am*np.sin(2*p2*x)
        x=np.linspace(0,t,nbE)
        y=psi(x)
        graph()
else:
    print("Mauvais numéro de signal")
```
